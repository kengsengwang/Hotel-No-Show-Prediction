```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, accuracy_score, roc_auc_score
```

```python
# Load dataset
def load_dataset(file_path):
    data = pd.read_csv(file_path)
    return data
```

```python
# Data Cleaning
def clean_data(data):
    # Convert 'price' to numeric after removing currency symbols
    data['price'] = data['price'].str.extract(r'(\d+\.\d+)').astype(float)

    # Replace 'num_adults' numeric words with integers
    num_adults_mapping = {'one': 1, 'two': 2}
    data['num_adults'] = data['num_adults'].replace(num_adults_mapping).astype(float)

    # Fill missing values in numerical columns with mean values
    numerical_cols = ['price', 'arrival_day', 'checkout_day', 'num_children']
    for col in numerical_cols:
        data[col] = data[col].fillna(data[col].mean())

    # Fill missing values in categorical columns with the most frequent value
    categorical_cols = ['branch', 'booking_month', 'arrival_month', 'checkout_month', 'country', 'first_time', 'platform', 'room']
    for col in categorical_cols:
        data[col] = data[col].fillna(data[col].mode()[0])

    return data
```

```python
# Perform EDA
def perform_eda(data):
    # Distribution of No-Shows
    plt.figure(figsize=(8, 5))
    sns.countplot(data=data, x='no_show')
    plt.title('Distribution of No-Shows')
    plt.xlabel('No-Show (0 = Showed, 1 = No-Show)')
    plt.ylabel('Count')
    plt.show()
```

```python
# Function to Encode Categorical Variables
from sklearn.preprocessing import LabelEncoder

def encode_categorical_features(data, categorical_columns):
    """
    Encodes categorical columns in the dataset using Label Encoding.

    Parameters:
        data (DataFrame): The dataset containing categorical columns to encode.
        categorical_columns (list): A list of column names to encode.

    Returns:
        DataFrame: The dataset with encoded categorical columns.
        dict: A dictionary of LabelEncoders for each encoded column.
    """
    label_encoders = {}
    for col in categorical_columns:
        le = LabelEncoder()
        data[col] = le.fit_transform(data[col].astype(str))
        label_encoders[col] = le
    return data, label_encoders
```

```python
# Function to Encode Categorical Variables
from sklearn.preprocessing import LabelEncoder

def encode_categorical_variables(data, categorical_columns):
    """
    Encodes categorical variables in the dataset using Label Encoding.

    Parameters:
        data (pd.DataFrame): The dataset containing the categorical columns.
        categorical_columns (list): A list of column names to encode.

    Returns:
        pd.DataFrame: The dataset with encoded categorical columns.
        dict: A dictionary containing the LabelEncoders for each encoded column.
    """
    label_encoders = {}
    for column in categorical_columns:
        le = LabelEncoder()
        # Convert the column to string type before encoding
        data[column] = le.fit_transform(data[column].astype(str))
        label_encoders[column] = le
    
    return data, label_encoders
```

```python
# Train and Evaluate Machine Learning Models
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

def train_and_evaluate_models(data, target_column='no_show', test_size=0.2, random_state=42):
    """
    Trains multiple machine learning models and evaluates their performance.

    Parameters:
        data (pd.DataFrame): The dataset containing features and the target column.
        target_column (str): The name of the target column. Default is 'no_show'.
        test_size (float): Proportion of the dataset to include in the test split. Default is 0.2.
        random_state (int): Random seed for reproducibility. Default is 42.

    Returns:
        dict: A dictionary of trained models.
        dict: A dictionary of train-test splits for features and target.
    """
    # Define features and target
    X = data.drop(columns=[target_column, 'booking_id'])
    y = data[target_column]

    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)

    # Define machine learning models
    models = {
        "Logistic Regression": LogisticRegression(max_iter=1000, random_state=random_state),
        "Random Forest": RandomForestClassifier(random_state=random_state),
        "Gradient Boosting": GradientBoostingClassifier(random_state=random_state)
    }

    # Return models and train-test data for further evaluation
    return models, {
        "X_train": X_train, 
        "X_test": X_test, 
        "y_train": y_train, 
        "y_test": y_test
    }
```

```python
# Train and evaluate each model
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report

def evaluate_models(models, X_train, X_test, y_train, y_test):
    """
    Trains and evaluates multiple models, calculating performance metrics.

    Parameters:
        models (dict): A dictionary of model names and their corresponding initialized model objects.
        X_train (pd.DataFrame): Training feature set.
        X_test (pd.DataFrame): Testing feature set.
        y_train (pd.Series): Training target set.
        y_test (pd.Series): Testing target set.

    Returns:
        dict: A dictionary containing evaluation metrics for each model.
    """
    results = {}

    for model_name, model in models.items():
        # Train the model
        model.fit(X_train, y_train)
        
        # Make predictions
        y_pred = model.predict(X_test)
        y_proba = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else None

        # Calculate metrics
        accuracy = accuracy_score(y_test, y_pred)
        roc_auc = roc_auc_score(y_test, y_proba) if y_proba is not None else None
        report = classification_report(y_test, y_pred, output_dict=True)

        # Store results
        results[model_name] = {
            "Accuracy": accuracy,
            "ROC AUC": roc_auc,
            "Classification Report": report
        }

    return results
```

```python
# Main Execution
if __name__ == "__main__":
    # Define file path
    file_path = "mini_project_2_data.csv"  # Update with the actual file path

    # Load and clean data
    print("Loading dataset...")
    data = load_dataset(file_path)
    print("Dataset loaded successfully.")

    print("Cleaning dataset...")
    data = clean_data(data)
    print("Dataset cleaned successfully.")

    # Perform EDA
    print("Performing Exploratory Data Analysis (EDA)...")
    perform_eda(data)
    print("EDA completed.")

    # Encode categorical columns
    categorical_cols = ['branch', 'booking_month', 'arrival_month', 'checkout_month', 'country', 'first_time', 'platform', 'room']
    print("Encoding categorical variables...")
    data, label_encoders = encode_categorical_variables(data, categorical_cols)
    print("Categorical variables encoded successfully.")

    # Train and evaluate models
    print("Training and evaluating models...")
    models, splits = train_and_evaluate_models(data, target_column='no_show')
    results = evaluate_models(models, 
                              X_train=splits["X_train"], 
                              X_test=splits["X_test"], 
                              y_train=splits["y_train"], 
                              y_test=splits["y_test"])
    print("Model training and evaluation completed.")

    # Print results
    for model_name, metrics in results.items():
        print(f"{model_name} Results:")
        print(f"Accuracy: {metrics['Accuracy']:.2f}")
        if metrics["ROC AUC"] is not None:
            print(f"ROC AUC: {metrics['ROC AUC']:.2f}")
        print("Classification Report:")
        print(metrics["Classification Report"])
        print("\n")
```
